```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_id_169484 = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
df_id_169523 = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
df_id_169524 = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
df_id_169529 = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
```

```python
df_id_169484.info()
df_id_169484.head()
```

```python
# Mapeo de números de mes a nombres
meses = {
    1: 'Enero',
    2: 'Febrero',
    3: 'Marzo',
    4: 'Abril',
    5: 'Mayo',
    6: 'Junio',
    7: 'Julio',
    8: 'Agosto',
    9: 'Septiembre',
    10: 'Octubre',
    11: 'Noviembre',
    12: 'Diciembre'
}


# Reemplazar los números de los meses en el DataFrame filtrado
df_id_169484['Numero Mes'] = df_id_169484['Numero Mes'].map(meses)

# Colores según el valor de la lluvia
colores = np.where(df_id_169484['Lluvia - mm'] > 20, 'red', 
                   np.where((df_id_169484['Lluvia - mm'] >= 10) & (df_id_169484['Lluvia - mm'] <= 20), 'yellow', 'lightgreen'))

# Crear la gráfica
plt.figure(figsize=(10,6))  # Tamaño de la figura
bars = plt.barh(df_id_169484['Numero Mes'], df_id_169484['Lluvia - mm'], color=colores)

# Añadir etiquetas de valor en cada columna
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}', 
             ha='left', va='center')
    
# Añadir etiquetas y título
plt.xlabel('Lluvia - mm')
plt.ylabel('Numero Mes')
plt.title(f'Lluvia mensual por ID 169484')
plt.show()
```
