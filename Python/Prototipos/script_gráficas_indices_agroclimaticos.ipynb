***Gráficas de Índices Agroclimáticos**

```python
import datetime as dt
import pandas as pd
import string as str
import numpy as np
import matplotlib.pyplot as plt
```

```python
conteo_dias_secos = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
conteo_dias_secos_por_mes = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
resultado_periodo_seco = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
cantidad_periodo_seco_mes = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
promedio_por_estacion = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
conteo_dias_humedos = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
conteo_dias_humedos_por_mes = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
resultado_periodo_humedo = pd.read_csv('.csv', header = 0,delimiter=';', encoding='latin-1')
```

```python
conteo_dias_secos.info()
conteo_dias_secos.head(50)
```

```python
# Convertir la columna 'ID' a objeto (texto)
conteo_dias_secos['ID'] = conteo_dias_secos['ID'].astype(object)
```

```python
# Ordenar el DataFrame por la columna Count_dias_secos de mayor a menor
conteo_dias_secos2 = conteo_dias_secos.sort_values(by='Count_dias_secos', ascending=False)
```

```python
conteo_dias_secos2.head()
```

```python
# Ordenar el DataFrame por la columna Count_dias_secos de mayor a menor
conteo_dias_secos2 = conteo_dias_secos2.sort_values(by='Count_dias_secos', ascending=False)

# Crear gráfica de barras con un estilo más atractivo
plt.figure(figsize=(20, 14))
bars = plt.bar(conteo_dias_secos2.index, conteo_dias_secos2['Count_dias_secos'], color='orange', edgecolor='black', linewidth=2)  # Añadir bordes a las barras

# Añadir título y etiquetas a los ejes
plt.title('Conteo de Días Secos por ID Por Ciclo de Cultivo', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('ID', fontsize=14)
plt.ylabel('Conteo de Días Secos', fontsize=14)

# Añadir etiquetas de valor en cada columna sin decimales y con un estilo mejorado
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{int(bar.get_height())}',  
             ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')  # Ajustar el estilo del texto

# Ajustar las etiquetas del eje x y mejorar el espaciado entre las barras
plt.xticks(conteo_dias_secos2.index, conteo_dias_secos2['ID'], fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()  # Mejorar el espaciado entre las etiquetas

# Personalizar el fondo y la cuadrícula de la gráfica
plt.gca().set_facecolor('lightgrey')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir una leyenda y personalizarla
plt.legend(['Conteo de Días Secos'], loc='upper right', fontsize=12)

# Guardar la gráfica como PNG
plt.savefig('C:/Users/stzor/Desktop/Junio/ICDF/Conteo_de_Días_Secos_por_ID_Por_Ciclo_de_Cultivo_11jun2024.png', format='png', dpi=300)
# Mostrar gráfica
plt.show()
```

conteo_dias_secos_por_mes

```python
#se podría hacer gráfica de líneas comparativa entre años cuando se tengan más datos
conteo_dias_secos_por_mes.info()
conteo_dias_secos_por_mes.head(10)
```

```python
import datetime

# Supongamos que year_actual es el año actual
year_actual = datetime.datetime.now().year

# Filtrar el DataFrame para el año actual
conteo_dias_secos_por_mes_filtrado = conteo_dias_secos_por_mes.loc[conteo_dias_secos_por_mes['Year'] == year_actual]

# Mostrar el DataFrame filtrado
print(conteo_dias_secos_por_mes_filtrado)
```

```python
conteo_dias_secos_por_mes_filtrado.info()
```

```python
# Obtener una lista única de meses y IDs
meses = conteo_dias_secos_por_mes_filtrado['Mes'].unique()
IDs = conteo_dias_secos_por_mes_filtrado['ID'].unique()
num_meses = len(meses)
num_IDs = len(IDs)

# Ancho de las barras
bar_width = 0.8 / num_IDs

# Crear gráfica de barras con un estilo más atractivo
plt.figure(figsize=(20, 14))

for i, mes in enumerate(meses):
    # Obtener los datos para el mes actual
    data_mes = conteo_dias_secos_por_mes_filtrado[conteo_dias_secos_por_mes_filtrado['Mes'] == mes]
    
    for j, ID in enumerate(IDs):
        # Obtener los datos para el ID actual
        data_ID = data_mes[data_mes['ID'] == ID]
        
        # Calcular la posición de la barra en el eje x
        x_position = i + j * bar_width
        
        # Crear la barra para el ID actual
        plt.bar(x_position, data_ID['Count_dias_secos_por_mes'], 
                width=bar_width, color=plt.cm.tab10(j), edgecolor='black', linewidth=2,
                label=f'ID {ID}')
        
        # Añadir etiquetas de valor en cada barra
        plt.text(x_position, data_ID['Count_dias_secos_por_mes'], f'{data_ID["Count_dias_secos_por_mes"].values[0]}',
                 ha='center', va='bottom', fontsize=10, color='black')

# Añadir título y etiquetas a los ejes
plt.title('Conteo de Días Secos por ID Por Mes', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Conteo de Días Secos por mes', fontsize=14)

# Ajustar las etiquetas del eje x
plt.xticks(np.arange(num_meses) + (0.8 / num_IDs) * (num_IDs - 1) / 2, meses, fontsize=12)

# Personalizar el fondo y la cuadrícula de la gráfica
plt.gca().set_facecolor('lightgrey')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir una leyenda y personalizarla
handles, labels = plt.gca().get_legend_handles_labels()
unique_labels = list(set(labels))
plt.legend(handles[:num_IDs], unique_labels[:num_IDs], loc='upper right', fontsize=12)


# Guardar la gráfica como PNG
plt.savefig('C:/Users/stzor/Desktop/Junio/ICDF/Conteo_de_Días_Secos_por_ID_mes_11jun2024.png', format='png', dpi=300)
# Mostrar gráfica
plt.show()
```
